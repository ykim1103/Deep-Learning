```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pylab as plt   # 그림그리기 위한 import
import sys
```

```python
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0], [0], [0], [1], [1], [1]]
```

```python
X=tf.placeholder(tf.float32,shape=[None,2])       # 6행 2열인데 6대신 None 이 나오는 이유는?
                                                   # 행의 갯수를 알아서 계산하게 하려고 None을 쓴다.


y=tf.placeholder(tf.float32,shape=[None,1])       # placeholder는 값을 고정시키기 위함?
#                                                    데이터타입이랑 형태가 고정인데  미리 주는게 아니라
#                                                    나중에 줄때     아직 데이터에 대해 모를때 주로 쓴대영.

 
    

W= tf.Variable(tf.random_normal([2,1]))           # 변수. 값이 바뀌는 랜덤값으로 지정 
b= tf.Variable(tf.random_normal([1]))


hypot=tf.sigmoid(tf.sigmoid(tf.matmul(X,W)+b))

cost= -tf.reduce_mean(y*tf.log(hypot)+(1-y)*tf.log(1-hypot))

train= tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(10000):
    sess.run(train,feed_dict={X:x_data,y:y_data})
```

```python
pred=tf.cast(hypot>0.5,dtype=tf.float32)
sess.run(pred,feed_dict={X:x_data,y:y_data})

accuracy=tf.reduce_mean(tf.cast(tf.equal(pred,y),dtype=tf.float32))
accuracy

h,p,a = sess.run([hypot,pred,accuracy],feed_dict={X:x_data,y:y_data})
print('가설: ', h, '\n예측: ', p, '\n정확도: ', a)
```

```python
X= tf.placeholder(tf.float32,shape=[None,16])    # 원래 열은 17개이지만 정답열 하나를 제외해서 16개로
y= tf.placeholder(tf.int32,shape=[None,1])     # 정답 열 1개만 
                                                # int32인 이유는 y값은 원핫인코딩을 이용해 0,1 값으로만 받을 것이기 때문

y_one_hot=tf.one_hot(y,7)    # 원핫인코딩

y_one_hot=tf.reshape(y_one_hot,[-1,7])


W= tf.Variable(tf.random_normal([16,7]))
b= tf.Variable(tf.random_normal([7]))       # W의 y값과 같으면 된다(?)


logit = tf.matmul(X,W)+b
hypot = tf.nn.softmax(logit)
cost = tf.nn.softmax_cross_entropy_with_logits(logits=logit,labels=y_one_hot)   




train= tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)




sess=tf.Session()
sess.run(tf.global_variables_initializer())

pred= tf.argmax(hypot,1)
correct=tf.equal(pred,tf.argmax(y_one_hot,1))
accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))




# for step in range(200):
#     sess.run(train,feed_dict={X:x_data,y:y_data})
    
    
#     if step % 10 ==0 :
#         sess.run([tf.reduce_mean(cost),accuracy],feed_dict={X:x_data,y:y_data})
        
#         print('step: {:5}\tLoss: {:.3f}\tAcc: {}'.fotmat(step,loss,acc))
    
for step in range(200):
    sess.run(train, feed_dict={X:x_data, y:y_data})
    
    if step % 10 == 0:
        loss, acc = sess.run([tf.reduce_mean(cost), accuracy], feed_dict={X:x_data, y:y_data}) # 비용 평균과 정확도
        print("step:{:5}\tLoss:{:.3f}\tAcc:{}".format(step, loss, acc))    
    
```

```python
# 데이터 가공 및 준비
x_data = xy[:, :-1]
y_data = xy[:, [-1]]

# 텐서 그래프 준비
X = tf.placeholder(tf.float32, shape=[None, 16])
y = tf.placeholder(tf.int32, shape=[None, 1])



# 원래
# W = tf.Variable(tf.random_normal([16, 1]))
# b = tf.Variable(tf.random_normal([1]))

# one_hot 매서드 활용하면?
y_one_hot = tf.one_hot(y, 7)
y_one_hot = tf.reshape(y_one_hot, [-1,7])

# 가중치와 바이어너리
W = tf.Variable(tf.random_normal([16,7]))
b = tf.Variable(tf.random_normal([7]))



# 가설설정
logit = tf.matmul(X,W) + b
hypot = tf.nn.softmax(logit)

# 비용함수
cost = tf.nn.softmax_cross_entropy_with_logits_v2(logits = logit, labels = y_one_hot)

# 최소비용게산
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

# 실행준비
sess = tf.Session()
sess.run(tf.global_variables_initializer())

# 성능 측정 준비
pred = tf.argmax(hypot, 1)                                # 예측
correct = tf.equal(pred, tf.argmax(y_one_hot, 1))         # 맞은 값 예측
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))   # 정확도


# 훈련 실행

for step in range(200):
    sess.run(train, feed_dict={X:x_data, y:y_data})
    
    if step % 10 == 0:
        loss, acc = sess.run([tf.reduce_mean(cost), accuracy], feed_dict={X:x_data, y:y_data}) # 비용 평균과 정확도
        print("step:{:5}\tLoss:{:.3f}\tAcc:{}".format(step, loss, acc))
```

### 사례2: MNIST 숫자이미지

```python
from tensorflow.examples.tutorials.mnist import input_data

mnist=input_data.read_data_sets('data/MNIST_data/',one_hot=True)
```

```python
type(mnist)
dir(mnist)
mnist.train.labels
mnist.train.num_examples   # 행개수 55000
```

```python
X = tf.placeholder(tf.float32,shape=[None,784])
y = tf.placeholder(tf.float32,shape=[None,10])

W = tf.Variable(tf.random_normal([784,10]))
b = tf.Variable(tf.random_normal([10]))

logit = tf.matmul(X,W)+b
hypot = tf.nn.softmax(logit)
cost = tf.nn.softmax_cross_entropy_with_logits(logits=logit,labels=y)

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

correct = tf.equal(tf.argmax(hypot, 1),tf.argmax(y,1))         # 맞은 값 예측
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))   # 정확도


sess=tf.Session()
sess.run(tf.global_variables_initializer())



batch_size=100

for epoch in range(15):
    total_batch = int(mnist.train.num_examples/batch_size)
    
    
    for i in range(total_batch):
        batch_xs,batch_ys = mnist.train.next_batch(batch_size)
        
        _, loss,acc = sess.run([train, tf.reduce_mean(cost),accuracy],feed_dict={X:batch_xs,y:batch_ys})
        
        print("step:{:5}\tLoss:{:.3f}\tAcc:{}".format(epoch, loss, acc))
```

```python
print("정확도 :", sess.run(accuracy, feed_dict={X:mnist.test.images, y:mnist.test.labels}))
```
